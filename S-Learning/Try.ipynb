```python
import matplotlib.pyplot as plt
import numpy as np
import scipy.stats as ss
```

```python
n = 100 # we create 100 random veriable
beta_0 = 5
beta_1 = 2

np.random.seed(1)
```

```python
x = ss.uniform.rvs(size=100) * 10
y = beta_0 + beta_1 * x + ss.norm.rvs(loc = 0, scale = 1, size=n)
```

```python
temp = ss.norm.rvs(loc = 0, scale = 1, size = 100)
```

```python
np.mean(temp)
```

```python
plt.figure()
plt.plot(x, y, "o", ms=5)
xx = np.array([0,10])
plt.plot(xx,beta_0 + beta_1 * xx)
plt.xlabel("x")
plt.ylabel("y")
```

```python
print(beta_0 + beta_1 * xx)
```

```python
# find RSS residual square of sum
rss = []
slops = np.arange(-10, 15, 0.01)
for s in slops:
    rss.append(np.sum((y - beta_0 - s*x)**2))
```

```python
min_ind = np.argmin(rss)
print("Estimate for slope: ", slops[min_ind])
```

```python
plt.figure()
plt.plot(slops, rss)
plt.xlabel("Slope")
plt.ylabel("RSS")
```
